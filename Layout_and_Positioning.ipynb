Question 1 : What are semantic HTML elements? Why is using them important for web
development?

Semantic HTML elements are HTML tags that clearly describe the meaning and purpose of the content they contain. They make the structure of a web page more understandable to browsers, search engines, and developers.

Common Sementic HTML Element:

Header:	Defines the header section of a page or section.

Nav:	Contains navigation links.

Main:	Represents the main content of the page.

Section:	Groups related content.

Article:	Represents independent content such as a blog post or news article.

Aside:	Contains supplementary content like sidebars.

Footer: Defines the footer section.

```python
<!DOCTYPE html>
<html>
<head>
    <title>My Blog</title>
</head>
<body>
    <header>
        <h1>My Blog</h1>
    </header>

    <nav>
        <a href="#">Home</a>
        <a href="#">About</a>
    </nav>

    <main>
        <article>
            <h2>Semantic HTML</h2>
            <p>Semantic elements give meaning to content.</p>
        </article>
    </main>

    <footer>
        <p>&copy; 2026 My Blog</p>
    </footer>
</body>
</html>
```

Why Semantic HTML Is Important:

1. Improves Accessibility

a)  Screen readers can better understand the page structure.


b) Helps users with disabilities navigate content more easily.

2. Improves Browser Understanding

a) Browsers can interpret content structure correctly.

b) Supports future web technologies and standard.

Question 2: You're designing a blog page. Which semantic elements would you use to
structure the page, and why?


```python
<!DOCTYPE html>
<html>
<head>
    <title>My Blog</title>
</head>
<body>

    <header>
        <h1>My Blog</h1>
        <p>Sharing knowledge and ideas</p>
    </header>

    <nav>
        <a href="/">Home</a>
        <a href="/blogs">Blogs</a>
        <a href="/about">About</a>
        <a href="/contact">Contact</a>
    </nav>

    <main>

        <article>
            <header>
                <h2>What is Semantic HTML?</h2>
                <p>Published on June 1, 2026</p>
            </header>

            <section>
                <h3>Introduction</h3>
                <p>Semantic HTML provides meaning to web content...</p>
            </section>

            <section>
                <h3>Benefits</h3>
                <p>It improves accessibility, SEO, and maintainability...</p>
            </section>

            <footer>
                <p>Author: John Doe</p>
            </footer>
        </article>

        <aside>
            <h3>Related Posts</h3>
            <ul>
                <li><a href="#">HTML Basics</a></li>
                <li><a href="#">CSS Layouts</a></li>
            </ul>
        </aside>

    </main>

    <footer>
        <p>&copy; 2026 My Blog. All rights reserved.</p>
    </footer>

</body>
</html>
```

1.Benefits of This Structure:

a) Accessibility:

. Screen readers can understand the page layout more easily.

b) SEO: Search engines can identify the main content (article) and page sections.

c) Maintainability: The code is easier for developers to read and update.

d) Meaningful Structure: Each element clearly communicates its role in the page.

Question 4: Identify and correct the errors in the following CSS code:
p {
 font-size: 16;
 color: #333
 margin-top 10px;
}

```python
p {
 font-size: 16;
 color: #333
 margin-top 10px;
}
```

Errors:

1. Missing unit for font-size:

16 should be 16px (or another valid unit).

2.Missing semicolon after color:

Question 6: Explain the CSS box model and its components.

The CSS Box Model is a fundamental concept in web design. It describes how every HTML element is represented as a rectangular box and how its size and spacing are calculated.

1. Content:

The actual content of the element, such as text, images, or other HTML elements.

```python
width: 300px;
height: 100px;
```

2. Padding:

The space between the content and the border.

```python
padding: 20px;
```

1. Increases the visible size of the element.

2. Background color extends into the padding area.

3. Border

The line surrounding the padding and content.

```python
border: 2px solid black;
```

. Can have different widths, styles, and colors.

4. Margin:

The space outside the border that separates the element from other element.

```python
margin: 15px;
```

1. Creates external spacing.
2. Is transparent.

Why the Box Model Is Important:

1.Controls element sizing and spacing.

2.Helps create consistent page layouts.

3.Essential for responsive web design.

4.Makes it easier to debug alignment and spacing issues.

Question 7: How do the relative, absolute, and fixed positioning properties differ in
CSS?

1. Relative Positioning (position: relative)

The element remains in the normal document flow but can be moved relative to its original position.

```python
.box {
    position: relative;
    top: 20px;
    left: 10px;
}
```

Characteristics:

1.Original space is still reserved.

2.Moved relative to its normal position.

3.Often used as a reference container for absolutely positioned child element.

Example:

```python
<div class="box">Relative Element</div>
```

2. Absolute Positioning (position: absolute)

The element is removed from the normal document flow and positioned relative to its nearest positioned ancestor (an ancestor with position: relative, absolute, fixed, or sticky).

```python
.container {
    position: relative;
}

.box {
    position: absolute;
    top: 20px;
    right: 10px;
}
```

Characteristics:

1.Removed from the normal page flow.

2.Other elements ignore its space.

3.Positioned relative to the nearest positioned parent.

4.If no positioned parent exists, it is positioned relative to the page (body).

Example:

```python
<div class="container">
    <div class="box">Absolute Element</div>
</div>
```

3. Fixed Positioning (position: fixed)

The element is removed from the normal flow and positioned relative to the browser viewport.

```python
.box {
    position: fixed;
    bottom: 20px;
    right: 20px;
}
```

Characteristics:

1.Stays in the same place even when the page is scrolled.

2.Positioned relative to the viewport, not a parent element.

3.Commonly used for navigation bars, chat buttons, and "Back to Top" button.

Example:

```python
<div class="box">Fixed Element</div>
```

Question 8: Write a CSS rule to set a background image for a (div) with the class
.banner, ensuring the image covers the entire area without repeating.

```python
.banner {
    background-image: url("banner.jpg");
    background-repeat: no-repeat;
    background-size: cover;
    background-position: center;
}
```

Explanation:

1.background-image: url("banner.jpg"); → Sets the background image.

2.background-repeat: no-repeat; → Prevents the image from repeating.

3.background-size: cover; → Scales the image to cover the entire <div> area while maintaining its aspect ratio.

4.background-position: center; → Centers the image within the (div).

Example HTML:

```python
<div class="banner">
    <h1>Welcome to My Website</h1>
</div>
```
